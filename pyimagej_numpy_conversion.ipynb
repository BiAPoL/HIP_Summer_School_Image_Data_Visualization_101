# pyimagej
Pyimagej is a python library that allows to run ImageJ2 algorithms from python.

See also
* [pyimagej setup / test](pyimagej.md)
* [pyimagej documentation](https://github.com/imagej/pyimagej/tree/master/doc)
* [pyimagej tutorial at I2K2020](https://github.com/imagej/i2k-2020-pyimagej)

```python
import imagej

# Start up ImageJ
ij = imagej.init()
```

Load example image from the web, e.g. using [scikit-image](https://scikit-image.org/):

```python
from skimage.io import imread
image = imread("https://fiji.sc/samples/blobs.png")
image
```

Convert the image from a numpy-array to ImageJ2:

```python
image_ij = ij.py.to_java(image)
image_ij
```

```python
# show the image
ij.py.show(image_ij)
```

Convert the image back to numpy

```python
image_np = ij.py.from_java(image_ij)
image_np
```

```python
from skimage.io import imshow

imshow(image_np)
```
